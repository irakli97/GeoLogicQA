```python
def summarize_results(question_answer_results):
    summary = {
        "total_questions": 0,
        "models": {}
    }

    # Initialize model counters
    model_names = set()
    for item in question_answer_results:
        if "results" in item:
            model_names.update(item["results"].keys())

    for model in model_names:
        summary["models"][model] = {
            "correct": 0,
            "total": 0
        }

    # Count correct and total answers per model
    for item in question_answer_results:
        if "results" in item:
            summary["total_questions"] += 1
            for model, is_correct in item["results"].items():
                summary["models"][model]["total"] += 1
                if is_correct:
                    summary["models"][model]["correct"] += 1

    # Calculate percentages
    for model in summary["models"]:
        correct = summary["models"][model]["correct"]
        total = summary["models"][model]["total"]
        percentage = (correct / total) * 100 if total > 0 else 0
        summary["models"][model]["percentage"] = round(percentage, 2)

    return summary

def print_results_table(question_answer_results):
    summary = summarize_results(question_answer_results)

    # Prepare table data
    headers = ["Model", "Correct", "Answered", "Accuracy (%)"]
    rows = []

    for model, data in summary["models"].items():
        rows.append([
            model,
            str(data["correct"]),
            str(data["total"]),
            f"{data['percentage']:.2f}%"
        ])

    # Find maximum column widths
    col_widths = [len(header) for header in headers]
    for row in rows:
        for i, cell in enumerate(row):
            col_widths[i] = max(col_widths[i], len(cell))

    # Create horizontal separator
    separator = "+" + "+".join(["-" * (w + 2) for w in col_widths]) + "+"

    # Print table
    print(f"\nEvaluation Results (Total Questions: {summary['total_questions']})")
    print(separator)

    # Print header
    header_row = "| " + " | ".join(
        header.ljust(col_widths[i]) for i, header in enumerate(headers)
    ) + " |"
    print(header_row)
    print(separator)

    # Print data rows
    for row in rows:
        data_row = "| " + " | ".join(
            cell.ljust(col_widths[i]) for i, cell in enumerate(row)
        ) + " |"
        print(data_row)

    print(separator)

print("kenguru")
print_results_table(question_answer_results)
print("komarovi")
print_results_table(komarovi)
```
